```python
# classificação
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

class classificacao:
  def mostrar_metricas(self, model, X_train, X_test, y_train, y_test):
    print('='*25)
    print('Teste')
    print('='*25)
    y_pred = model.predict(X_test)
    print(accuracy_score(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print('='*25)
    print('Treino')
    print('='*25)
    y_pred = model.predict(X_train)
    print(accuracy_score(y_train, y_pred))
    print(confusion_matrix(y_train, y_pred))

  def logistic(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'penalty': 'l2', # 'l1', 'l2', 'elasticnet', 'none'
      'dual': False,
      'tol': 0.0001,
      'C': 1.0,
      'fit_intercept': True,
      'intercept_scaling': 1,
      'solver': 'lbfgs', # 'newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga', 'newton-cholesky', None
      'max_iter': 100,
      'multi_class': 'auto', # 'ovr', 'multinomial', 'auto'
      'verbose': 0,
      'warm_start': False,
      'n_jobs': None,
      'l1_ratio': None
    }
    """
    clf = LogisticRegression(**kwargs, random_state=42)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf
  def svc(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'C': 1.0,
      'kernel': 'rbf', # 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
      'degree': 3,
      'gamma': 'scale', # 'scale', 'auto'
      'coef0': 0.0,
      'shrinking': True,
      'probability': False,
      'tol': 0.001,
      'cache_size': 200,
      'max_iter': -1,
      'decision_function_shape': 'ovr', # 'ovo', 'ovr'
      'break_ties': False,
    }
    """
    clf = SVC(**kwargs, random_state=42)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf

  def sgd(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'loss': 'hinge', # 'log', 'modified_huber', 'squared_hinge', 'perceptron', 'squared_loss', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'
      'penalty': 'l2', # 'l1', 'l2', 'elasticnet', 'none'
      'alpha': 0.0001,
      'l1_ratio': 0.15,
      'fit_intercept': True,
      'max_iter': 1000,
      'tol': 0.001,
      'shuffle': True,
      'epsilon': 0.1,
      'n_jobs': None,
      'learning_rate': 'optimal', # 'constant', 'optimal', 'invscaling', 'adaptive'
      'eta0': 0.0,
      'power_t': 0.5,
      'early_stopping': False,
      'validation_fraction': 0.1,
      'n_iter_no_change': 5,
      'class_weight': None,
      'warm_start': False,
      'average': False
    }
    """
    clf = SGDClassifier(**kwargs, random_state=42)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf

  def knn(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'n_neighbors': 5,
      'weights': 'uniform', # 'distance', 'uniform'
      'algorithm': 'auto', # 'ball_tree', 'kd_tree', 'brute', 'auto'
      'leaf_size': 30,
      'p': 2,
      'n_jobs': None
    }
    """
    clf = KNeighborsClassifier(**kwargs)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf

  def arvore(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'criterion': 'gini', # 'gini', 'entropy', 'log_loss'
      'splitter': 'best', # 'best', 'random'
      'max_depth': None,
      'min_samples_split': 2,
      'min_samples_leaf': 1,
      'min_weight_fraction_leaf': 0.0,
      'max_features':None, # int, float or {sqrt, log2},
      'max_leaf_nodes': None,
      'min_impurity_decrease': 0.0,
      'ccp_alpha': 0.0,
      'monotonic_cst': None
    }
    """
    clf = DecisionTreeClassifier(**kwargs, random_state=42)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf

  def floresta(self, X_train, X_test, y_train, y_test, **kwargs):
    """
    param = {
      'n_estimators': 100,
      'criterion': 'gini', # 'gini', 'entropy', 'log_loss'
      'max_depth': None,
      'min_samples_split': 2,
      'min_samples_leaf': 1,
      'min_weight_fraction_leaf': 0.0,
      'max_features':None, # int, float or {sqrt, log2},
      'max_leaf_nodes': None,
      'min_impurity_decrease': 0.0,
      'ccp_alpha': 0.0,
      'monotonic_cst': None,
      'bootstrap': True,
      'oob_score': False,
      'n_jobs': None,
      'verbose': 0,
      'warm_start': False,
      'class_weight': None,
      'max_samples': None,
    }
    """
    clf = RandomForestClassifier(**kwargs, random_state=42)
    clf.fit(X_train, y_train)
    self.mostrar_metricas(clf, X_train, X_test, y_train, y_test)
    return clf
```
